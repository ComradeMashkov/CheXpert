```python
import os
import pandas as pd
import numpy as np
from PIL import Image
```

```python
classes = ['No Finding', 'Enlarged Cardiomediastinum', 'Cardiomegaly', 'Lung Opacity', 
           'Lung Lesion', 'Edema', 'Consolidation', 'Pneumonia', 'Atelectasis', 'Pneumothorax', 
           'Pleural Effusion', 'Pleural Other', 'Fracture', 'Support Devices']
```

```python
parent_directory = 'train_classified'
for cls in classes:
    directory = cls
    path = os.path.join(parent_directory, cls)
    os.mkdir(path)
    print(f'Папка \'{directory}\' создана')
```

```python
classes_dict = {}
classes_count = np.zeros(len(classes))
for index, cls in enumerate(classes):
    classes_dict[cls] = classes_count[index]
```

```python
df = pd.read_csv('train.csv')
df.head(6)
```

```python
for index, rows in df.iterrows():
    path_to_img = rows['Path'][20:]
    for cls in classes:
        if rows[cls] == 1 and rows['Frontal/Lateral'] == 'Frontal':
            img = Image.open(path_to_img)
            img_dir = f'train_classified/{cls}/{cls}_{int(classes_dict[cls])}.jpg'
            img.save(img_dir)
            classes_dict[cls] += 1
print('Готово.')
```
